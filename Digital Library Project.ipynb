## Simple Object Recognition using color seperation
For this simple case, we assume that we only and one type of coins and bills. The coins will be represented as circles with red color, and the bills will be represented as rectangles with blue color. The input image is shown below.

### Seperate Objects
The way we extract each bills and coins is applying thresh hold. But just applying reguare threshold algorithm will not distinguish between the bills and coins because the objects in the output image will be just binary. The way we handle this situation is using **HSV** color representaion of the image. The **HSV** stands for **Hue-Saturation-Value**, and the colors of the image will be separated into ranges. 

    - Hue : This range will vary from 0 degree to 360 degree which start from red(0 ~ 120) to green(120 ~ 240) and so on.
    - Saturation : This represent how deep the color is. The range will be from 0 to 255.
    - Value : This is the lightness of the color which vary from 0 to 255.
    
Since our bills have color blue and coins have color red, the **Hue** range for bills will be from 240 to 360 degree and coins will be from 0 to 120.

We first convert the image with HSV color, and then set the lower and upper bounds for each color range and make mask using **cv2.inRange()** in order to threshold each objects. The thresholding will be done using **cv2.bitwise_and()** operator.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import skimage
import cv2
import numpy as np

img = cv2.imread('sample1.png')
plt.imshow(img), plt.show()
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# range for bills.
low_bill = np.array([0, 50, 0])
high_bill = np.array([100, 255, 255])

# range for coins.
low_coin = np.array([120, 50, 0])
high_coin = np.array([240, 255, 255])

# make masks for each bill and coin.
mask_bill = cv2.inRange(hsv_img, low_bill, high_bill)
mask_coin = cv2.inRange(hsv_img, low_coin, high_coin)

# Extract bills and coins using logical operator.
bill = cv2.bitwise_and(hsv_img, hsv_img, mask=mask_bill)
coin = cv2.bitwise_and(hsv_img, hsv_img, mask=mask_coin)

# plot.
plt.subplot(1,2,1)
plt.title('Bills')
plt.imshow(bill)
plt.subplot(1,2,2)
plt.title('Coins')
plt.imshow(coin)
plt.show()
```

### Find boundries
After we seperate each objects(bills, coins), we can find the boundaries of each object using **findContours()**. The purpose of this step is to crop each coin and bill. For this simple objects, we might not need to crop each object but if each object is complicated, then this step is neccessary for classification process.

The function **cv2.findContours()** assumes that the input image is binary image. So we need to threshold the object extracted image. From the HSV image, the *V* channel can be used for grayscale image. So we take this channel for grayscale input image to **cv2.threshold()** function. Once we get the binary image, we can find the boundaries of each object.

```python
# find contours.
ret_bill,thresholded_bill = cv2.threshold(bill[:,:,2],0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
contours_bill, hier_bill = cv2.findContours(thresholded_bill, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
ret_coin,thresholded_coin = cv2.threshold(coin[:,:,2],0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
contours_coin, hier_coin = cv2.findContours(thresholded_coin, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# draw contours.
cont_draw_bill = np.zeros(thresholded_bill.shape)
cv2.drawContours(cont_draw_bill, contours_bill, -1, (255, 0, 0), 3)
cont_draw_coin = np.zeros(thresholded_coin.shape)
cv2.drawContours(cont_draw_coin, contours_coin, -1, (255, 0, 0), 3)
plt.subplot(1,2,1)
plt.title('Bills')
plt.imshow(cont_draw_bill, cmap='gray')
plt.subplot(1,2,2)
plt.title('Coins')
plt.imshow(cont_draw_coin, cmap='gray')
plt.show()
```

### Extract each object
Once we found each boundary of the objects, we can use **cv2.boundingRect()** function to find the region of rectangle that enclose the object. This function returns x, y, w, h where x and y are the pixel postions, and w and h are the width and height of the rectangle. After we find the regions, we can extract each object from the original input image.

```python
# find enclosing rectangles.
# bills.
num_contour = len(contours_bill)
indices = range(num_contour)
for contour, index in zip(contours_bill, indices):
    x, y, w, h = cv2.boundingRect(contour)
    plt.subplot(1, num_contour, index+1)
    plt.title(str(index+1) + 'th bill')
    plt.imshow(img[y:y+h, x:x+w])
plt.show()

# coins.
num_contour = len(contours_coin)
indices = range(num_contour)
for contour, index in zip(contours_coin, indices):
    x, y, w, h = cv2.boundingRect(contour)
    plt.subplot(1, num_contour, index+1)
    plt.title(str(index+1) + 'th bill')
    plt.imshow(img[y:y+h, x:x+w])
plt.show()
```

# General Object Recognition
Object recognition from general image requires similar steps but more complicated. For simple shape, finding object region using contours is simple task in the above case because the thresholded image is clear, and we assumed that each object has distinct color. But for general objects, this is not the case so there should be problem finding clear contour around the object.

We have a image of bills and coins as shown below. If we just apply same method as above, the extracted contour has many holes in it which implies that it is hard to extract the object because we have no idea which contour is real object. So, in this case, the goal should be finding right contour around the real object so that we can find box region around the object. In order to do that, filling those holes inside bigger contour will make each object as one region. In this way, we can extract the right objects.

```python
# read image.
img = cv2.imread('sample2.png', 0)
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')

# threashold the image.
ret,th = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.subplot(1,2,2)
plt.title('Thresholded Image')
plt.imshow(th, cmap='gray')
plt.show()

kernel = np.ones((10,10))
open_morph = cv2.morphologyEx(th, cv2.MORPH_OPEN, kernel)
plt.imshow(open_morph, cmap='gray'), plt.show()

# find contours.
cnt, hier = cv2.findContours(open_morph, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
temp_img = np.zeros(open_morph.shape)

cv2.drawContours(temp_img, cnt, -1, 255, 3)
plt.title('Contour Image')
plt.imshow(temp_img, cmap='gray')
plt.show()
```

### Filling inner contours
One way to filling the inner contour is using **drawContours()** with -1 for thickness argument. If we pass the thickness argument with negative value, it will fill out the inner area as opencv api document says. For thresholding, it seems not good if we apply Otsu thresholding to filling inner contour area. Instead of using threshold method, it was better to invert the image using **cv2.bitwise_not()** and filling the area.

```python
# invert = cv2.bitwise_not(img)
# plt.imshow(invert, cmap='gray'), plt.show()
img = cv2.imread('sample2.png', 0)
invert = cv2.bitwise_not(img) # invert the image.
cnt, hier = cv2.findContours(invert, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
cnt_img = np.zeros(invert.shape)
cv2.drawContours(cnt_img, cnt, -1, 255, -1)
plt.subplot(1,2,1)
plt.imshow(cnt_img, cmap='gray')

# remove holes in the circle.
kernel = np.ones((5,5))
close_morph = cv2.morphologyEx(cnt_img, cv2.MORPH_CLOSE, kernel)
close_morph = close_morph.astype(np.uint8)
plt.subplot(1,2,2)
plt.imshow(close_morph, cmap='gray')
plt.show()


# find contours from clean image.
objects = []
cnt, hier = cv2.findContours(close_morph, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# find bounding rectangles.
for each in cnt:
    x, y, w, h = cv2.boundingRect(each)
    temp_img = img[y:y+h, x:x+w] # y and x are reversed.
    objects += [temp_img]

# display.
i = 1
for obj in objects:
    plt.subplot(1,len(cnt),i)
    plt.title(str(i) + "th object")
    plt.imshow(obj, cmap='gray')
    i += 1
plt.show()
```

### Matching
For feature extraction algorithm, we are going to use SIFT. Since SIFT algorithm calculates feature vectors from the local patch around the keypoint, it would be better choice for our project because it will be robust to geometrical changes of the objects in the image. 

```python
# get the sample image(dollar bill, coin)
dollar = cv2.imread('dollar1.png', 0)
five_dollar = cv2.imread('five.jpeg', 0)
quarter = cv2.imread('coin.png', 0)
sample_images = [dollar, five_dollar, quarter]

# find descriptors from the part we extracted and from the sample image.
sift = cv2.SIFT()
sample_descriptors = []
for sample_object, index in zip(sample_images, range(len(sample_images))):
    kp, des = sift.detectAndCompute(sample_object, None)
    sample_descriptors += [(index, des)]

# For extracted parts.
object_descriptors = [] # descriptors extracted from the objects in the input image.
for obj, index in zip(objects, range(len(objects))):
    kp, des = sift.detectAndCompute(obj, None)
    # set_des = [(kp1, des1, index1), (kp2, des2, index2), ...]
    object_descriptors += [(kp, des, index)]

# set up the matching conditions, and instantiate matcher.
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks = 50)
flann = cv2.FlannBasedMatcher(index_params, search_params)


# perform matches.

def match_object(sample_descriptors, object_descriptors):
    """
    This function takes sample_descriptors which contains the training image index, and its descriptor, and returns
    a list that consists triples where each triple has training image index, extracted object image index, and the
    number of descriptors match.
    """
    # sample_descriptors = (sample_objects_index, feature_descriptor)
    # object_descriptors = [(kp, des, index), ...]
    num_matches = []
    for each in object_descriptors:
        matches = flann.knnMatch(sample_descriptors[1], each[1], k=2) # each[1] is obj descriptor.
        matched = []
        for m,n in matches:
            # small distance is the better. 
            if m.distance < 0.7*n.distance:
                matched.append(m)
        # num_matches += [(sample_object_index, object index, number of matches), ...]
        num_matches += [(sample_descriptors[0], each[2], len(matched))]
        
    # sort by the number of matches... more matches, similar to the sample object.
    sorted(num_matches, key=lambda x : -x[2]) # num_matches[i][2] is the number of matches.
    # return best matches.
    # We say that if there are more than 10 matches between sample object and candidate object,
    # they are same objects.
    good_matches = []
    for each in num_matches:
        if each[2] >= 20:
            good_matches += [each]
    return good_matches


def count(sample_descriptors):
    """
    This function takes sample_descriptors which consist of descriptors extracted from training image and the index of the
    image. It returns a dictionary that contains the number of object found for each category.
    """
    amount = {"Dollar" : [], "Five" : [], "Quarter" : []}
    # check in order of dollar -> five dollar -> quarter
    for descriptor, object_name in zip(sample_descriptors, ["Dollar", "Five", "Quarter"]):
        good_matches = match_object(descriptor, object_descriptors)
        # sample_descriptors[0] is the sample object index.
        # amount[object] = [number_of_descriptor_matches, sample_object_index, [all index of extracted objects]]
        # One thing to notice is that good_matches[2] is the number of mathced descriptors from match_object().
        amount[object_name] = [good_matches[0][2], descriptor[0], [object_index for sample_object_index, object_index, num in good_matches]] 
    return amount

amount = count(sample_descriptors)
for each in amount:
    # check two conditions. If there were object match and number of matches is one or more.
    if len(amount[each]) > 0 and amount[each][0] >= 1:
        # if there is actual match, show all mached objects.
        # since amount[object] = [number_of_descriptor_matches, sample_object_index, [all index of extracted objects], 
        # and we only need the sample_object_index and object_index(extracted objects), we set sample_index = sample_object_index
        # and object_index = object_index from extracted object image index. 
        # Here, amount[object][2] is a list that contains object_index.
        for sample_object_index, object_index in zip([amount[each][1] for i in range(len(amount[each][2]))], amount[each][2]):
            plt.subplot(1, 2, 1)
            plt.title("Sample Image")
            plt.imshow(sample_images[sample_object_index], cmap='gray')
            plt.subplot(1,2,2)
            plt.title("Extracted Object") # each is object name. Dollar or Five.
            plt.imshow(objects[object_index], cmap='gray')
            plt.show()
        
```
